```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
# Load data
data = np.genfromtxt('Real_estate.csv', delimiter=',')
data = data[1:] # xÃ³a header
data = data[:, 1:]
print(data.shape)

X = data[:, :-1]
y = data[:, -1]
print(X.size)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
data_train = np.append(X_train, np.expand_dims(y_train, axis=1), axis=1)
```

```python
#transaction date - price correlation graph
trans_date = X_train[:, 0]
price = y_train


plt.scatter(trans_date, price)
plt.grid()

plt.xlabel("trans_date")
plt.ylabel("price")

slope, intercept = np.polyfit(trans_date, price, 1)

def yfit(x):
    return slope*x + intercept

xfit = np.linspace(min(trans_date), max(trans_date), 1000)
plt.plot(xfit, yfit(xfit), color = 'r')
```
